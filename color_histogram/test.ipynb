```python
import glob
from PIL import Image
import color_histogram
import numpy as np
```

```python
def test():
    image_list = []
    percentage_list = []
    colors_list = []
    filename_list = []

    for index, path in enumerate(glob.glob('/Users/cemkilinc/Desktop/test_out/*.png')): #assuming gif
        img=Image.open(path)
        temp = img.copy()
        img.close()
        img = temp
        colors, percentage= color_histogram.get_image_color_features(img)
        filename = path[path.rfind("/")+1:]
        filename_list.append(filename)
        image_list.append(img)
        percentage_list.append(percentage)
        colors_list.append(colors)

    image_list = np.array(image_list, dtype=object)
    percentage_list = np.array(percentage_list, dtype=object)
    colors_list = np.array(colors_list, dtype=object)
    return filename_list, colors_list, percentage_list
```

```python
filename_list, colors_list, percentage_list = test()
```

```python
filee = "image2sub1.png"
i = filename_list.index(filee)
color_arr_1 = colors_list[i]
percentage_1 = percentage_list[i]

```

```python
number_of_images = len(filename_list)
score_arr = []
for j in range(0, number_of_images):
        #print(str(i) + " " + str(j))

        
        color_arr_2 = colors_list[j]
        percentage_2 = percentage_list[j]

        score = color_histogram.find_similiarity_score(color_arr_1, percentage_1, color_arr_2, percentage_2)
        print(filee + " and " + filename_list[j] + " similarity score: " + str(score))
        score_arr.append(score)
```

```python
numpi = np.array(score_arr)
argu = np.argsort(numpi)
for arg in argu:
    print(filename_list[arg])   
```
